```python
import numpy as np
endings = np.array(["st","nd","rd"])

letterList = np.arange(0,26)
for aLetter in letterList:
    currLetter = chr(aLetter+ord('a'))
    if( aLetter%10>2 or 20>aLetter>9 ):
        print("%s is the %gth letter of the alphabet."%(currLetter,aLetter+1))
    else:
        print("%s is the %g%s letter of the alphabet."%(currLetter,aLetter+1,endings[aLetter%10]))
```

```python
x = np.linspace(-5,5,101)
y = x**3
plt.plot(x,y,label='x^3')
plt.xlim([x[0],x[-1]])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('example plot')
plt.ylim(min(y),max(y))
plt.legend()
plt.gca().set_aspect(1/plt.gca().get_data_ratio())
plt.show()
```

```python
#make a purple triangle

x = np.linspace(0,10,101)

rangeOfPlots = np.arange(0,1,.02)

for aSeed in rangeOfPlots:
    colorGrad = (aSeed/max(rangeOfPlots),.5/max(rangeOfPlots),aSeed/(2*max(rangeOfPlots)))
    plt.plot(x,x**aSeed, color=colorGrad)
    plt.plot(-x,x**aSeed, color=colorGrad)
    plt.plot(x,-x**aSeed+20, color=colorGrad)
    plt.plot(-x,-x**aSeed+20, color=colorGrad)
plt.gca().set_aspect(1/plt.gca().get_data_ratio())
plt.annotate('Hello',(-1.2,10))
plt.show()
```

```python
x = np.linspace(-4.5,4.5,11)
y = x**2 + np.cos(x) *10
minValTup =  (x[np.argsort(y)[0]],min(y))
minValTup
plt.plot(x,y,'ko-',markeredgecolor='k',markerfacecolor='purple',markersize=12)
plt.annotate('Hello',minValTup ,xytext = (-1.2,14),arrowprops=
             {'width':3}
            )
plt.show()
```
