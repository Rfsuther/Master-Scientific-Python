```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000

x = np.zeros(N)
y = np.zeros(N)
xNxt = 0
yNxt = 0
for i in range(1,N):
    k = np.random.randint(1,4)
    y[i] = y[i-1]/2
    x[i]  = x[i-1]/2 + k-1
    if k==2:
        y[i] += 2
             
plt.plot(x,y,'k.')
plt.axis('off')
plt.show()
```
