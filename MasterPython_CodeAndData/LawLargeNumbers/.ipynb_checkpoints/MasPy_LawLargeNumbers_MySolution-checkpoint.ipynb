```python
# load modules
import numpy as np
import matplotlib.pyplot as plt
```

# Generate a population of random numbers

```python
# simulation parameters

population_size = 2.3e5

# 50 samples of 500 each
sample_size = 50
number_of_samples = 500

rng = np.random.default_rng()
```

```python
# generate population distribution

# a population of numbers that decreases with increasing value
allPeople = np.logspace(np.log10(.001),np.log10(5),int(population_size))
population = 1/allPeople
#population = 

# compute the true population mean
trueMean = np.mean(population)
 
# plot the data
plt.subplot(1,2,1)
plt.plot(np.arange(0,len(population[::1000])),population[::1000])
plt.xlabel('Data value')
plt.ylabel('Sample')

# shuffle the data values
plt.subplot(1,2,2)
np.random.shuffle(population)
plt.scatter(np.arange(0,len(population[::1000])),population[::1000])
plt.show()
```

# Monte Carlo sampling

```python
# generate one random sample from the population

np.random.choice(population)
```

```python
# Monte Carlo sampling for sample means
def MonteCarlo(population,number_of_samples,sample_size):
    
    samplemeans = np.zeros(number_of_samples)
    for expi in range(number_of_samples):
        samplemeans[expi] = np.mean(np.random.choice(population,size=sample_size,replace=True))
    return samplemeans


samplemeans = MonteCarlo(population,number_of_samples,sample_size)
```

```python
# plot the sample means, and a line showing the true mean


plt.xlabel('Sample number')
plt.ylabel('Mean value')
plt.scatter(np.arange(number_of_samples),samplemeans,c='k',label='Sample means')
plt.plot(np.arange(number_of_samples),np.ones(number_of_samples)*trueMean,'r',label='True mean')
plt.legend()
plt.show()
```

# Cumulative averaging

```python

# Cumulative averaging
# the ith value in vector cumave is the mean of all previous points
# do this first in a loop
cumAve1 = np.zeros(number_of_samples)
cumAve2 = np.zeros(number_of_samples)

for i in range(number_of_samples):
    cumAve1[i] = np.sum(samplemeans[0:i+1])/(i+1)

# then try to get the same result without a loop!
cumAve2 = np.cumsum(samplemeans)/(np.arange(number_of_samples)+1)

# plot the results, along with a horizontal line showing the true mean
plt.xlabel('Sample number')
plt.ylabel('Mean value')
plt.plot(np.arange(number_of_samples),cumAve1,'ko',label='cumAve1')
plt.plot(np.arange(number_of_samples),cumAve2,'b+',label='cumAve2')
plt.plot(np.ones(number_of_samples)*trueMean,'r-',label='cumAve2')

plt.legend()
plt.show()
```

```python
# compute the squared divergence from the true mean, 
# repeated over 100 iterations

import time

numRuns = 100
runs = np.zeros([numRuns,number_of_samples])
averageDenominator = (np.arange(number_of_samples)+1)

for i in range(runs.shape[0]):
    currentRun = MonteCarlo(population,number_of_samples,sample_size)
    runs[i,:] = np.cumsum(currentRun)/averageDenominator
```

```python
# plot them all
for i in range(runs.shape[0]):
    plt.plot(abs(runs[i,:]-trueMean)**2)
plt.xlabel('Sample number')
plt.ylabel('Squared divergence from true mean')
plt.ylim([-10,500])
plt.show()
```
